### Data Preprocessing

```python
# removing columns that have no descriptions in the original data source
# these two columns have same values across all rows
data = data.drop(["Z_CostContact", "Z_Revenue"], axis=1)
```

```python
# converting 'Dt_Customer' variable to datetime type
data["Dt_Customer"] = pd.to_datetime(data["Dt_Customer"], format="%d-%m-%Y")
```

```python
# creating a new variable to track the # days since customer joined, with respect to the maximum joining date
max_joining_date = data["Dt_Customer"].max()
data["Days_Customer"] = max_joining_date - data["Dt_Customer"]
data["Days_Customer"] = data["Days_Customer"].dt.days.astype(int)
```

```python
# feature engineering
# renaming column (confusing name - Response is a binary variable of whether a customer accepted the last offer)
data.rename(columns = {"Response" : "AcceptedCmp6"}, inplace = True)
# customer age as of current year 2023
data["Age"] = 2023-data["Year_Birth"] 
data = data.drop(["Dt_Customer", "Year_Birth"], axis=1)
# total amount spent on products
data["Total_Spent"] = data["MntWines"] + data["MntFruits"] + data["MntMeatProducts"] + \
data["MntFishProducts"] + data["MntSweetProducts"] + data["MntGoldProds"]
# total number of items purchased
data["Total_NumPurchase"] = data["NumDealsPurchases"] + data["NumWebPurchases"] + data["NumCatalogPurchases"] + data["NumStorePurchases"] 
# total number accepting offers
data["Total_Offers"] = data["AcceptedCmp1"] + data["AcceptedCmp2"] + data["AcceptedCmp3"] + data["AcceptedCmp4"] + \
data["AcceptedCmp5"] + data["AcceptedCmp6"]
```

```python
# Income variable imputation
# checking distribution
print('Skew :', round(data["Income"].skew(), 2))
fig, ax = plt.subplots()
data["Income"].hist(bins=50)
ax.grid(False)
plt.show()
```

Since Income variable is positively skewed, I'll replace the missing values with the median as median is less sensitive to outliers than the mean.

```python
# Income variable imputation
data["Income"] = data["Income"].fillna(data["Income"].median())
data["Income"].describe()
```

After imputing missing values for 'Income' variable with median, I checked how the statistics changed. Previously with missing values, the mean was around 52247 and standard deviation was around 25173. The mean value didn't change much but the standard deviation decreased slightly. 	

```python
# looking into values of categorical variables
# groupings are rather non-informative and can become more compact
print(data["Marital_Status"].value_counts(), "\n")
print(data["Education"].value_counts())
```

```python
# modifying categorical variables
# since 'Alone', 'Absurd', 'YOLO' are all singles they can be grouped together; their respective counts are also not high
data["Marital_Status"] = data["Marital_Status"].replace({"Alone":"Single","Absurd":"Single","YOLO":"Single", \
                                                         "Together":"Partner","Widow":"Widowed"})
data["Education"] = data["Education"].replace({"Graduation":"Bachelor","Basic":"Bachelor","2n Cycle":"Master"})
```

```python
data.describe(include="all", datetime_is_numeric=True)
```

### EDA - Univariate Analysis

**Categorical Variables**

```python
# Education variable 
Ed_cat = data["Education"].value_counts().index
Ed_count = data["Education"].value_counts().values
# Marital Status variable 
Mar_cat = data["Marital_Status"].value_counts().index
Mar_count = data["Marital_Status"].value_counts().values

fig, arr = plt.subplots(1, 2, figsize=(15,7))
colors = sns.color_palette('pastel')

arr[0].bar(Ed_cat, Ed_count, color=colors)
arr[0].set_xlabel("Education")
arr[0].set_ylabel("Count")

arr[1].bar(Mar_cat, Mar_count, color=colors)
arr[1].set_xlabel("Marital Status")
arr[1].set_ylabel("Count")
plt.show()
```

**Numerical Variables Outliers**

```python
# extract numeric columns only
numeric_cols = data.select_dtypes(include=np.number)
numeric_cols = numeric_cols.drop(columns=['ID', 'Kidhome', 'Teenhome','AcceptedCmp1', 'AcceptedCmp2', 'AcceptedCmp3', \
                           'AcceptedCmp4', 'AcceptedCmp5', 'AcceptedCmp6', 'Complain', 'Total_Offers'])

# create subplots
numeric_cols.plot(subplots=True, layout=(5,5), kind='box', figsize=(12,16), patch_artist=True)
plt.suptitle('Find Outliers', fontsize=15, y=0.9)
plt.subplots_adjust(wspace=0.5)
```

While most variables seem to have natural outliers (e.g. 'MntWines'), 'Year_Birth' and 'Age' (derived from 'Year_Birth') seem to be errors since it is impossible for a person borned in 1900s to be still alive. I'll drop the outliers from dataset.

```python
data = data[(data["Age"]<100)]
print("Total number of rows after removing outliers:", len(data))
```

```python
plt.figure(figsize=(20,20))

heatmap_data = data[["Income", "Kidhome", "Teenhome", "Recency", "Total_NumPurchase", "Total_Spent", "Days_Customer", \
                    "Age", "Total_Offers"]]
sns.heatmap(heatmap_data.corr(numeric_only = True), annot = True, \
            vmin = -1, vmax = 1, linewidth=.5, cmap = 'RdBu')
plt.show()
```

Several high-level observations can be made from the correlation matrix above. 
- 'Income' is positively correlated with both 'Total_Spent' and 'Total_NumPurchase', which is rather intuitive.
- 'Kidhome' (binary variable indicating whehther there's a kid in the household) is negatively correlated with 'Total_Spent' and 'Total_NumPurchase', which I first thought as unexpected at first; however, it might be the case that households with kids tend to save more thus spending less. 
- 'Total_Spent' and 'Total_NumPurhcase' are positively correlated - which is also intuitive and provides a sanity check.
